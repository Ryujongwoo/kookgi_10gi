```python
# 선택 정렬의 반복 알고리즘을 이용해 내림차순(큰 점수가 1등) 석차를 계산하는 프로그램
score = [75, 94, 89, 95, 62]
# 석차를 기억할 리스트를 만든다. => 석차를 기억하는 기억장소의 초기치는 무조건 1로 한다.
# rank = [1, 1, 1, 1, 1]
# rank = []
# for i in range(len(score)):
    # rank.append(1)
rank = [1 for i in range(len(score))]
# print(rank)

for i in range(len(score) - 1):
    for j in range(i + 1, len(score)):
        # i번째 점수가 크면 j번째 석차를 1증가 시키과 j번째 점수가 크면 i번째 석차를 1증가 시킨다.
        # 부등호 방향을 모두 반대로 변경하면 오름차순(작은 점수가 1등) 석차가 계산된다.
        if score[i] > score[j]:
            rank[j] += 1
        elif score[i] < score[j]:
            rank[i] += 1
            
for i in range(len(score)):
    print('{0:3d}점은 {1}등 입니다.'.format(score[i], rank[i]))
```

```python
# 점수와 석차를 출력하고 별점을 부여한다. => 10점 단위로 까만별 1개씩, 1자리 점수가 5점 이상이면 하얀별을 출력한다.
for i in range(len(score)):
    print('{0:3d}점은 {1}등 입니다. '.format(score[i], rank[i]), end = '')
    print('★' * (score[i] // 10), end = '')
    # 1자리 점수가 5점 이상이면 '☆'을 출력한다.
    # if score[i] % 10 >= 5:
        # print('☆', end = '')
    # print()
    print('☆' if score[i] % 10 >= 5 else '')
```

```python
# 점수에 따른 세로 별점
for i in range(10, 0, -1):
    print('{0:3d} │ '.format(i * 10), end = '')
    
    for s in score:
        if s // 10 >= i:
            print('  ★  ', end = '')
        elif s + 5 >= i * 10:
            print('  ☆  ', end = '')
        else:
            print('     ', end = '')
    print()
    
print('  0 └', end = '')
print('─' * 26)
print('      ', end = '')
for i in score:
    print(' {0:3d} '.format(i), end = '')
```
