피보나치 수열의 n번째 항의 값은 n-2 번째 항의 값과 n-1 번째 항의 값을 더한 값이다.  
P = 1 + 1 + 2 + 3 + 5 + 8 + 13 + 21 + ...

키보드로 피보나치 수열의 합계를 계산할 항(n)을 입력받아 n번째 항 까지 피보나채 수열의 합계를 계산해 출력하는 프로그램  
1번째 항과 2번째 항의 합계는 이미 계산되어있다고 가정하고 3번째 항 부터 n번째 항 까지 반복하며 수열의 합계를 계산한다.

```python
a = 1 # 첫 번째 항의 값 초기화, 피보나치 수열의 n-2 번째 항의 값
b = 1 # 두 번째 항의 값 초기화, 피보나치 수열의 n-1 번째 항의 값
y = 2 # 첫 번째 항의 값과 두 번째 항의 값을 더한 합계 => n 번째 항 까지 피보나치 수열의 합계를 기억할 변수

# 피보나치 수열의 합계를 계산할 항(n)을 입력받는다.
n = int(input('피보나치 수열의 합계를 계산할 항을 입력하세요 : '))

# 1번째 항과 2번째 항은 이미 계산이 되어있다는 가정하에 작업을 하므로 3번째 한 부터 n번째 항 까지 반복하며 합계를 계산한다.
for k in range(3, n + 1):
    c = a + b
    y += c
    # 다음 항의 값을 계산하기 위해 n-2 번째 항의 값(a)을 n-1 번째 항의 값(b)으로 넣어주고 n-1 번째 항의 값(b) n번째 항의 값(c)을 넣어준다.
    a = b
    b = c
# ===== for k 끝
print('피보나치 수열 {} 번째 항 까지의 합계 : {}'.format(n, y))
```

```python
p = [1, 1] # 피보나치 수열의 첫 번째 항과 두 번째 항의 값을 기억하는 리스트를 만든다.
y = 2
n = int(input('피보나치 수열의 합계를 계산할 항을 입력하세요 : '))
for k in range(3, n + 1):
    p.append(p[k - 3] + p[k - 2])
    y += p[k - 1]
print(p, y)

print('피보나치 수열 {} 번째 항 까지의 합계 : {} = '.format(n, y), end = '')
for i in range(len(p) - 1):
    print(p[i], end = ' + ')
print(p[-1])

print('피보나치 수열 {} 번째 항 까지의 합계 : {} = '.format(n, y), end = '')
for i in range(len(p)):
    if i > 0:
        print(' + ', end = '')
    print(p[i], end = '')
print()
    
for i in range(len(p)):
    p[i] = str(p[i])
print('피보나치 수열 {} 번째 항 까지의 합계 : {} = {}'.format(n, y, ' + '.join(p)), end = '')
```
