유클리드 호제법 => 최대공약수, 최소공배수  
숫자 2개를 입력받아 큰수와 작은수를 구분한다.  
큰수를 작은수로 나눈 나머지가 0이면 작은수가 최대공약수가 되고 큰수를 작은수로 나눈 나머지가 0이 아니면 큰수를 기억하던 기억장소에 작은수를 넣어주고 작은수를 기억하던 기억장소에 나머지를 넣어준 후 작업을 반복한다.  
최대공약수가 계산되면 입력받은 숫자 2개를 곱한 결과를 최대공약수로 나눠주면 최소공배수를 얻을 수 있다.

```python
from IPython.display import Image
Image('./euclid_1.png', width = '650')
```

```python
# 숫자 2개를 입력받아 큰수, 작은수를 판단한다.
a, b = map(int, input('숫자 2개를 입력하세요 : ').split())
if a > b:
    big = a
    small = b
else:
    big = b
    small = a

while True:
    r = big % small # 큰수를 작은수로 나눈 나머지를 계산한다.
    if r == 0:      # 무한 루프 종료 조건
        break       # 무한 루프를 탈출한다.
    # ===== if r == 0: 끝
    # 큰수를 작은수로 나눠 떨어지지 않았으므로 큰수에 작은수를 작은수에 나머지를 넣어준다.
    big = small
    small = r
# ===== while True: 끝

# 최소공배수를 계산하고 최대공약수와 최소공배수를 출력한다.
print('최대공약수 : {}, 최소공배수 : {}'.format(small, int(a * b / small)))
```

```python
a, b = map(int, input('숫자 2개를 입력하세요 : ').split())
r = 1
if a > b:
    high = a
    low = b
else:
    low = a
    high = b
    
while r > 0:
    r = high % low
    high = low
    low = r
# ===== while r > 0: 끝

l = int(a * b / high)
print('최대공약수 : {}, 최소공배수 : {}'.format(high, l))
```
