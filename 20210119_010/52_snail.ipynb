```python
a = [[0] * 5 for i in range(5)]
n = 0  # 1 ~ 25  1  
s = 1  #   1  -1     
i = 0  # 
j = -1 # 
k = 5  #     

while True:
    #    .
    for p in range(1, k + 1):
        n += 1
        j += s
        a[i][j] = n
    # ===== for k 
    k -= 1
    if k <= 0:
        break
    #    .   
    for p in range(1, k + 1):
        n += 1
        i += s
        a[i][j] = n
    # ===== for k 
    s *= -1
# ===== while True 

for i in range(len(a)):
    for j in range(len(a[0])):
        print('{0:3d} '.format(a[i][j]), end = '')
    print()
```
