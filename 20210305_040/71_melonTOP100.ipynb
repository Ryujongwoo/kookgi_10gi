```python
import warnings
warnings.filterwarnings(action = 'ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
targetSite = 'https://www.melon.com/chart/index.htm'
request = requests.get(targetSite)
print(request) # <Response [406]> => 헤더 문제로 사이트의 요청한 정보를 얻어올 수 없다.
```

```python
targetSite = 'https://www.melon.com/chart/index.htm'
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
request = requests.get(targetSite, headers = header)
print(request)
```

```python
html = request.text
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('div', {'class': 'ellipsis rank01'})
# print(titles)
# print(len(titles))

for title in titles:
    # print(title.text.strip())
    # print(title.text.split('\n')[1])
    print(title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
artists = soup.findAll('span', {'class': 'checkEllipsis'})
# print(artists)
# print(len(artists))

i = 0
for artist in artists:
    i += 1
    # print(i, artist.text.strip())
    # print(i, artist.text.split('\n')[0])
    print(i, artist.text.strip().split('\n')[0])
```

```python
print(dt.now())
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{0:3d}위 {1} - {2}'.format(i + 1, artist, title))
```

```python
file = open('melonTOP100.txt', 'w')
file.write('{} 현재 Melon 뮤직 실시간 TOP 100\n'.format(dt.now()))
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    data = '{0:3d}위 {1} - {2}'.format(i + 1, artist, title)
    file.write(data + '\n')   
file.close()
print('melonTOP100.txt 파일로 쓰기 완료')
```

```python
try:
    file = open('melonTOP100.txt', 'r')
    lines = file.readlines()
    for line in lines:
        print(line.strip())
except FileNotFoundError:
    print('디스크에 파일이 없습니다.')
```
