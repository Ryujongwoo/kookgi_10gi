```python
import warnings
warnings.filterwarnings(action = 'ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
print(dt.now())
```

```python
targetSite = 'https://music.bugs.co.kr/chart'
request = requests.get(targetSite)
# print(request)
html = request.text
soup = BeautifulSoup(html, 'html.parser')
# print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('p', {'class': 'title'})
# print(titles)
# print(len(titles))
for title in titles:
    # print(title.text.strip())
    # print(title.text.split('\n')[1])
    print(title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
artists = soup.findAll('p', {'class': 'artist'})
# print(artists)
# print(len(artists))
i = 0
for artist in artists:
    i += 1
    print(i, artist.text.strip().split('\n')[0])
```

```python
print(dt.now())
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{0:3d}위 {1} - {2}'.format(i + 1, artist, title))
```

```python
# 크롤링한 결과를 텍스트 파일에 저장한다.
file = open('bugsTOP100.txt', 'w')
# UnicodeEncodeError: 'cp949' codec can't encode character '\xf1' in position 29: illegal multibyte sequence
# 크롤링 데이터를 텍스트 파일에 저장하려는 경우 위와 같이 한글 인코딩 문제로 UnicodeEncodeError가 발생될 수 있다.
# 에러가 발생되면 file = open('bugsTOP100.txt', 'w', -1, 'UTF-8') 형태로 파일을 open 한다.

file.write('{} 현재 Bugs 뮤직 실시간 TOP 100\n'.format(dt.now()))
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    data = '{0:3d}위 {1} - {2}'.format(i + 1, artist, title)
    file.write(data + '\n')
    
file.close()
print('bugsTOP100.txt 파일로 쓰기 완료')
```

strftime('출력서식') 함수를 이용해서 날짜와 시간에 서식을 지정할 수 있다.

```python
# %Y => 년도 4자리, %y => 년도 2자리, %m => 월(숫자), %B => 월(영어 전체), %b(%h) => 월(영어 3글자), %d => 일
# %A => 요일(영어 전체), %a => 요일(영어 3글자)
now = dt.now()
print(now)
print(now.strftime('%Y-%m-%d'))
print(now.strftime('%Y.%m.%d'))
print(now.strftime('%Y/%m/%d'))
# 날짜 및 시간 서식에 한글을 사용하면 유니코드 문제로 인해서 아래와 같은 에러가 발생한다.
# UnicodeEncodeError: 'locale' codec can't encode character '\ub144' in position 2: Illegal byte sequence
# print(now.strftime('%Y년 %m월 %d일')) # 에러 발생
print(now.strftime('%Y년 %m월 %d일'.encode('unicode-escape').decode()).encode().decode('unicode-escape'))
print(now.strftime('%y-%B-%d'))
print(now.strftime('%y-%b-%d'))
print(now.strftime('%A %B-%d-%Y'))
print(now.strftime('%a %d. %h %Y'))
```

```python
# %H => 시간(24시각), %I => 시간(12시각), %M => 분, %S => 초, %p => AM/PM, %f => 마이크로 초
print(now.strftime('%H:%M:%S'))
print(now.strftime('%p %I:%M:%S.%f'))
print(now.strftime('%H시 %M분 %S초'.encode('unicode-escape').decode()).encode().decode('unicode-escape'))
```

```python
file = open('bugsTOP100.txt', 'w')
file.write('{} 현재 Bugs 뮤직 실시간 TOP 100\n'.format(dt.now().strftime('%Y년 %m월 %d일 %H시 %M분 %S초'.encode(\
                                'unicode-escape').decode()).encode().decode('unicode-escape')))
for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    data = '{0:3d}위 {1} - {2}'.format(i + 1, artist, title)
    file.write(data + '\n')    
file.close()
print('bugsTOP100.txt 파일로 쓰기 완료')
```

```python
# with 구문을 사용해 파일을 open 하면 with 내부의 내용이 모두 실행되고 난 후 자동으로 파일을 close 시켜준다.
with open('bugsTOP100.txt', 'w') as file:
    file.write('{} 현재 Bugs 뮤직 실시간 TOP 100\n'.format(dt.now().strftime('%Y년 %m월 %d일 %H시 %M분 %S초'.encode(\
                                'unicode-escape').decode()).encode().decode('unicode-escape')))
    for i in range(100):
        artist = artists[i].text.strip().split('\n')[0]
        title = titles[i].text.strip().split('\n')[0]
        data = '{0:3d}위 {1} - {2}'.format(i + 1, artist, title)
        file.write(data + '\n')
print('bugsTOP100.txt 파일로 쓰기 완료')
```

```python
# 텍스트 파일에 저장된 데이터를 읽어서 화면에 출력한다.
try:
    file = open('bugsTOP100.txt', 'r')
    # readlines() 메소드는 텍스트 파일의 내용 전체를 한꺼번에 읽어서 리스트로 리턴한다.
    lines = file.readlines()
    for line in lines:
        # print(line, end = '')
        print(line.strip())
except FileNotFoundError:
    print('디스크에 파일이 없습니다.')
```
