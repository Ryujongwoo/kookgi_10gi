```python
import warnings
warnings.filterwarnings(action = 'ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
import time
```

```python
# 타겟 사이트 주소 만들기
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210305&hh=11&rtm=Y&pg=1 => 1 ~50
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210305&hh=11&rtm=Y&pg=2 => 51 ~ 100
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210305&hh=11&rtm=Y&pg=3 => 101 ~ 150
# https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210305&hh=11&rtm=Y&pg=4 => 151 ~ 200
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210305&hh=11&rtm=Y&pg='
for i in range(1, 5):
    # targetSite = url + str(i)
    targetSite = '{}{}'.format(url, i)
    print(targetSite)
```

```python
# 노래 제목 크롤링
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210305&hh=11&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers = header)
    # print(request)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    # print(soup)
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    # print(titles)
    # print(len(titles))
    
    for title in titles:
        # print(title.text.strip())
        # print(title.text.split('\n')[-1].strip())
        print(title.text.strip().split('\n')[0])
```

```python
# 아티스트 크롤링
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210305&hh=11&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 2):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers = header)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')

    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    # print(artists)
    # print(len(artists))
    
    i = 0
    for artist in artists[5:]:
        i += 1
        # print(i, artist.text.strip())
        # print(i, artist.text.split('\n')[0])
        print(i, artist.text.strip().split('\n')[0])
    print('=' * 80)
```

```python
# 크롤링한 데이터를 리스트에 저장한다.
genie = []
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20210305&hh=11&rtm=Y&pg='
header = {'User-agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko'}
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers = header)
    time.sleep(1)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    
    for j in range(50):
        genie.append(
            (titles[j].text.strip().split('\n')[0], artists[j].text.strip().split('\n')[0])
        )
# print(len(genie))
print(genie)
```

```python
print(dt.now())
for i in range(200):
    print('{0:3d}위 {1} - {2}'.format(i + 1, genie[i][1], genie[i][0]))
```

```python
file = open('genieTOP200.txt', 'w')
file.write('{} 현재 Genie 뮤직 실시간 TOP 100\n'.format(dt.now()))
for i in range(200):
    data = '{0:3d}위 {1} - {2}'.format(i + 1, genie[i][1], genie[i][0])
    file.write(data + '\n')   
file.close()
print('genieTOP200.txt 파일로 쓰기 완료')
```
