```python
import warnings
warnings.filterwarnings(action = 'ignore')
from bs4 import BeautifulSoup
from datetime import datetime as dt
from selenium import webdriver
```

```python
# selenium을 사용해 가상 크롬을 실행한다. => 사용중인 크롬 버전과 같은 버전의 크롬 드라이버를 다운받는다.
driver = webdriver.Chrome('./chromedriver.exe')
# get() 메소드로 가상 크롬에 크롤링 할 타겟 사이트를 띄운다.
driver.get('https://www.nate.com/')
html = driver.page_source
# print(html)
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
issues = soup.findAll('a', {'class': 'ik'})
for issue in issues:
    print(' '.join(issue.text.split()[:-1]), end = ' ')
    check = issue.text.split()[-1]
    if check[:2] == '상승':
        print('↑' + check[2:])
    elif check[:2] == '하락':
        print('↓' + check[2:])
    elif check[:2] == '동일':
        print('-')
    else:
        print(check)
```
